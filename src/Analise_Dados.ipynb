# Análise dos Dados


```python
import pandas as pd

%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

## A. Preparação

```python
data = '../data/campaigns.csv'
df   = pd.read_csv(data, encoding="utf-8", sep=",")
```

```python
df.columns = ['index', 'age', 'job', 'marital', 'education', 'default',
       'balance', 'housing', 'loan', 'contact', 'day', 'month', 'duration',
       'campaign', 'pdays', 'previous', 'poutcome', 'y']
```

## 1. Analisando Clientes Positivos  / Clientes Negativos 

Clientes Positivos. y - 'yes', Clientes Negativos y = 'no'

```python
df_yes = df.copy()
df_yes = df_yes[df_yes['y']=='yes']

df_no = df.copy()
df_no = df_no[df_no['y']=='no']
```

### 1.1 Clientes Positivos/Negativos - Profissões

```python
sns.set(rc={'figure.figsize':(18,5)})
job_count = df_yes['job'].value_counts()

sns.set(style="darkgrid")
sns.barplot(job_count.index, job_count.values, alpha=0.9)
plt.title('Clientes Positivos - Profissões')
plt.ylabel('Número de Ocorrências', fontsize=12)
plt.xlabel('Profissões', fontsize=12)
plt.show()


job_count = df_no['job'].value_counts()

sns.set(style="darkgrid")
sns.barplot(job_count.index, job_count.values, alpha=0.9)
plt.title('Clientes Negativos - Profissões')
plt.ylabel('Número de Ocorrências', fontsize=12)
plt.xlabel('Profissões', fontsize=12)
plt.show()
```

```python
sns.set(rc={'figure.figsize':(18,8)})
labels = df_yes['job'].astype('category').cat.categories.tolist()
counts = df_yes['job'].value_counts()

sizes = [counts[var_cat] for var_cat in labels]
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True) 
ax1.axis('equal')
plt.show()


sns.set(rc={'figure.figsize':(18,8)})
labels = df_no['job'].astype('category').cat.categories.tolist()
counts = df_no['job'].value_counts()

sizes = [counts[var_cat] for var_cat in labels]
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True) 
ax1.axis('equal')
plt.show()
```

### 1.2 Clientes Positivos/Negativos - Estado Civil

```python
labels = df_yes['marital'].astype('category').cat.categories.tolist()
counts = df_yes['marital'].value_counts()
sizes = [counts[var_cat] for var_cat in labels]
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True) #autopct is show the % on plot
ax1.axis('equal')
plt.show()

labels = df_no['marital'].astype('category').cat.categories.tolist()
counts = df_no['marital'].value_counts()
sizes = [counts[var_cat] for var_cat in labels]
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True) #autopct is show the % on plot
ax1.axis('equal')
plt.show()
```

### 1.3 Clientes Positivos / Negativos  - Profissão / Rendimentos

```python
df_yes.boxplot('balance','job',rot = 30,figsize=(15,6))
df_no.boxplot('balance','job',rot = 30,figsize=(15,6))
```
