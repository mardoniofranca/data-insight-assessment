# Segmentação dos Clientes


```python
import pandas as pd
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
from sklearn.preprocessing import LabelEncoder
```

## A. Preparação

```python
data = '../data/campaigns.csv'

df   = pd.read_csv(data, encoding="utf-8", sep=",")

df.columns = ['index', 'age', 'job', 'marital', 'education', 'default',
       'balance', 'housing', 'loan', 'contact', 'day', 'month', 'duration',
       'campaign', 'pdays', 'previous', 'poutcome', 'y']
```

```python
lb_make = LabelEncoder()
df['job_code'] = lb_make.fit_transform(df['job'])
df['marital_code'] = lb_make.fit_transform(df['marital'])
df['education_code'] = lb_make.fit_transform(df['education'])
df['default_code'] = lb_make.fit_transform(df['default'])
df['housing_code'] = lb_make.fit_transform(df['housing'])
df['loan_code'] = lb_make.fit_transform(df['loan'])
df['contact_code'] = lb_make.fit_transform(df['contact'])
df['poutcome_code'] = lb_make.fit_transform(df['poutcome'])
```

```python
df.columns
```

```python
cols  = ['age', 'balance', 'duration',  'job_code', 'marital_code',
        'education_code', 'default_code', 'housing_code', 'loan_code',
        'contact_code', 'poutcome_code', 'y']
cols_f = ['age', 'balance', 'duration',  'job_code', 'marital_code',
        'education_code', 'default_code', 'housing_code', 'loan_code',
        'contact_code', 'poutcome_code']
```

```python
df = df[[]]
```

```python
df.head()
```

### 1. Análise Baseada em Clusterização para Clientes Positivos  / Clientes Negativos 

```python
df_yes = df.copy()
df_yes = df_yes[df_yes['y']=='yes']

df_no = df.copy()
df_no = df_no[df_no['y']=='no']
```

```python
print(len(df_yes), len(df_no))
```

```python
sns.kdeplot(df_yes['balance'].dropna())
```

```python
sns.kdeplot(df_no['balance'].dropna())
```

```python
sns.countplot(df_yes['marital_code'])
```

```python
sns.countplot(df_no['marital_code'])
```

```python
sns.lineplot('balance', 'age', data=df_yes)
```

```python
sns.lineplot('balance', 'age', data=df_no)
```

```python
from sklearn.cluster import KMeans
```

```python
df_yes = df_yes[cols_f] 
```

```python
df_no = df_no[cols_f] 
```

```python
df_yes.dtypes
```

```python
distortions = []
K = range(1,10)
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(df_yes)
    distortions.append(kmeanModel.inertia_)
```

```python
distortions_no = []
K = range(1,10)
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(df_no)
    distortions_no.append(kmeanModel.inertia_)
```

```python
plt.figure(figsize=(12,5))
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
```

```python
plt.figure(figsize=(12,5))
plt.plot(K, distortions_no, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
```

```python
kmeanModel = KMeans(n_clusters=5)
kmeanModel.fit(df_yes)
```

```python
kmeanModel_no = KMeans(n_clusters=5)
kmeanModel_no.fit(df_no)
```

```python
df_yes['k_means']=kmeanModel.predict(df_yes)
```

```python
df_no['k_means']=kmeanModel.predict(df_no)
```

```python
df_yes.head()
```

```python
df_no.head()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16,8))

axes[0].scatter(df_yes.balance, df_yes.age, c=df_yes['k_means'], cmap=plt.cm.Set1)
axes[0].set_title('K_Means Clientes Positivos ', fontsize=18)

axes[1].scatter(df_no.balance, df_no.age, c=df_no['k_means'], cmap=plt.cm.Set1)
axes[1].set_title('K_Means Clientes Negativos ', fontsize=18)
```

### comentário final: 
1.  Os agrupamentos (Clientes Positicos e Clientes Negativos), a principio apresenta números 
    de grupos, de acordo com o Método Elbow entre 5 e 6 grupos
    
2.  Aplicamos o k-means para identificar os grupos.

2.1 No caso de Clientes Positivos, os grupos são mais homogêneos em relação aos Clientes Negativos
